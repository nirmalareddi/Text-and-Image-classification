```python
#load the dataset and import libraries

import numpy as np
import matplotlib.pyplot as plt

X = np.load("/content/sample_data/unlabelled_train_data_images.npy")
```

```python
#get the shape and dimension of the dataset

print("Shape of image data:", X.shape)
print("Data type:", X.dtype)
```

```python
#Visualize sample of 25 images

plt.figure(figsize=(6, 6))
for i in range(25):
    plt.subplot(5, 5, i + 1)
    plt.imshow(X[i].squeeze(), cmap='gray')
    plt.axis("off")
plt.suptitle("Sample Images from Dataset", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# Step 1: Remove the extra dimension
X = X.reshape(X.shape[0], 28, 28)  # from (60000, 1, 28, 28) to (60000, 28, 28)

# Step 2: Flatten and normalize
X_flat = X.reshape(X.shape[0], -1)  # (60000, 784)
X_flat = X_flat.astype('float32') / 255.0

# Check result
print("New shape after flattening:", X_flat.shape)
print("Pixel range: ", X_flat.min(), "to", X_flat.max())
```

```python
from sklearn.decomposition import PCA

# Apply PCA
pca = PCA(n_components=50, random_state=42)
X_pca = pca.fit_transform(X_flat)

# Check new shape
print("Shape after PCA:", X_pca.shape)

# Plot cumulative explained variance
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.plot(np.cumsum(pca.explained_variance_ratio_), marker='o')
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("Explained Variance by PCA Components")
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import KMeans

# Step 4: KMeans clustering
kmeans = KMeans(n_clusters=10, random_state=42, n_init=10)
Y_train = kmeans.fit_predict(X_pca)

# Check shape and a sample
print("Shape of Y_train:", Y_train.shape)
print("Unique cluster labels assigned:", np.unique(Y_train))
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Reduce to 2D for plotting
pca_2d = PCA(n_components=2)
X_2d = pca_2d.fit_transform(X_flat)

plt.figure(figsize=(8, 6))
scatter = plt.scatter(X_2d[:, 0], X_2d[:, 1], c=Y_train, cmap="tab10", alpha=0.6, s=1)
plt.title("2D PCA Visualization of Clusters")
plt.colorbar(scatter, ticks=range(10))
plt.show()
```

```python
#Pick the top 10000 datapoints and apply tsne. The resultant plot should have different labels and numbers for the respective clusters. Make sure the colours are distinct

import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
import seaborn as sns

# Assuming X_flat and Y_train are already defined from the previous code

# Pick the top 10000 data points
X_subset = X_flat[:10000]
Y_subset = Y_train[:10000]

# Apply t-SNE
tsne = TSNE(n_components=2, random_state=42, perplexity=30, n_iter=300) # Increased n_iter for better convergence
X_tsne = tsne.fit_transform(X_subset)


# Create the plot
plt.figure(figsize=(10, 8))
sns.scatterplot(x=X_tsne[:, 0], y=X_tsne[:, 1], hue=Y_subset, palette="deep", legend="full") # Using seaborn's distinct palette

# Customize the plot for better readability
plt.title("t-SNE Visualization of Clusters (Top 10000 Data Points)")
plt.xlabel("t-SNE Component 1")
plt.ylabel("t-SNE Component 2")

# Annotate some points for clarity (optional, comment out if not needed)
#for i in range(0, len(X_tsne), 50):  # Adjust the step for fewer annotations
#  plt.annotate(Y_subset[i], (X_tsne[i, 0], X_tsne[i, 1]))

plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

# Train-test split
X_train, X_val, y_train, y_val = train_test_split(X_pca, Y_train, test_size=0.2, random_state=42, stratify=Y_train)

# Train Logistic Regression
clf = LogisticRegression(max_iter=1000)
clf.fit(X_train, y_train)

# Predict and evaluate
y_pred = clf.predict(X_val)
acc = accuracy_score(y_val, y_pred)

print(f"Validation Accuracy: {acc:.4f}")
print("\nClassification Report:")
print(classification_report(y_val, y_pred))
```

```python
# Evaluating with different classifiers for KMeans clusters.

from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import LinearSVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, classification_report

# Reuse your train-test split
models = {
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
    "Linear SVC": LinearSVC(max_iter=10000, random_state=42),
    "KNN": KNeighborsClassifier(n_neighbors=5),
    "MLP Neural Net": MLPClassifier(hidden_layer_sizes=(128,), max_iter=300, random_state=42)
}

results = {}

for name, model in models.items():
    print(f"\n Training {name}...")
    model.fit(X_train, y_train)
    y_pred = model.predict(X_val)
    acc = accuracy_score(y_val, y_pred)
    report = classification_report(y_val, y_pred, output_dict=True)
    results[name] = {
        "accuracy": round(acc, 4),
        "macro avg F1-score": round(report["macro avg"]["f1-score"], 4),
        "weighted avg F1-score": round(report["weighted avg"]["f1-score"], 4)
    }
    print(f" {name} - Accuracy: {acc:.4f}")
```
